```python
# Importing the Dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Data Collection and Data Processing
#loading the dataset to a pandas Dataframe
sonar_data = pd.read_csv('/Users/xiaosihuang/jupyter/porject1/sonar_rock_mine_data.csv', header=None)
```

```python
# Display the first 5 rows of the dataset
sonar_data.head()
```

```python
# number of rows and columns
sonar_data.shape
```

```python
# Generate statistical measures of the dataset (mean, std, min, max, etc.)
sonar_data.describe()
```

```python
# Count occurrences of each unique value in column 60
sonar_data[60].value_counts()
```

```python
# Group by column 60 and calculate the mean for each group
sonar_data.groupby(60).mean()
```

```python
# Remove column 60 and set the remaining columns as features (X)
# Extract column 60 as the target variable (Y)
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
# Print the feature data X (dataset without the target column)
print(X)
# Print the target variable Y (labels in column 60)
print(Y)
```

```python
# Training and Test data
```

```python
# Split the dataset into training and testing sets (10% for testing), stratified by Y
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
# Print the dimensions of X, X_train, and X_test
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Print the feature data of the training set (X_train)
print(X_train)
```

```python
# Print the target variable of the training set (Y_train)
print(Y_train)
```

```python


#Model Training --> Logistic Regression
```

```python
# Create an instance of the Logistic Regression model
model = LogisticRegression()
```

```python
#training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

```python


# Model Evaluation
```

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) 
```

```python
# Print the model's accuracy on the training data
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test) 
```

```python
# Print the model's accuracy on the test data
print('Accuracy on test data : ', test_data_accuracy)
```
